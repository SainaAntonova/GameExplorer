```python
import requests
import json
import time
from datetime import datetime
```

```python
APPDETAILS_URL = "https://store.steampowered.com/api/appdetails/"
```

```python
def get_app_details2(app_id, language='english'):
    params = {
        'appids': app_id,
        'cc': 'us',
        'l': language
    }
    response = requests.get(APPDETAILS_URL, params=params)
    if response.status_code == 200:
        app_details = response.json()
        return app_details
    else:
        print(f"Failed to get details for app_id: {app_id}")
        return None
```

```python
import os
def save_game_data(game_data, filename='games_data.txt'):
    filepath = os.path.join('/content/drive/MyDrive/steam_pars', filename)
    with open(filepath, 'a') as file:
        json.dump(game_data, file)
        file.write('\n')
```

```python
def get_steam_info(ids, range_start):
    game_count = 0
    for i in range(len(ids)):
        game_count += 1
        id_data = get_app_details2(ids[i], language='english')
        save_game_data(id_data, f'{str(range_start)}_{str(range_start+1000)}.txt')
        print(i, ids[i])
        time.sleep(1.5) # Если будет падать с ошибкой, как-то связанной с ответом сервера, можно попробовать поменять значение на 2
        if game_count % 1000 == 0:
            range_start += 1000
```

```python
range_start, range_end = 15000, 35546
get_steam_info(ids[range_start:range_end], range_start)
```
